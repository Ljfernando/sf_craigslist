```python
import pandas as pd
import json
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df[df.county=='San Francisco'].describe()
```

```python
sf = df[df.county=='San Francisco']
sf.shape
```

```python
sf['title'] = [title.split('\n')[0] for title in sf.title.values]
sf = sf.loc[sf.title.str.contains('\(')]
sf['neighborhood'] = [nh.split('(')[1][:-1] for nh in sf.title.values]
sf['title'] = [nh.split('(')[0] for nh in sf.title.values]
```

```python
valid_neighborhoods = sf.neighborhood.value_counts()[sf.neighborhood.value_counts() > 4].index
valid_n = np.asarray([curr in valid_neighborhoods for curr in sf.neighborhood.values])
sf = sf.loc[valid_n]
```

```python
sf['price_bucket'] = pd.qcut(sf.price, 4, labels=[0,1,2,3])
```

```python
sf.price_bucket.value_counts()
```

```python
sf.shape
```

```python
sf.to_csv('sf.csv', index=False)
```
